```python
import numpy as np
import seaborn as sns 
import pandas as pd
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore")
df=pd.read_csv("IPL.csv")
```

```python
# check the size of rows and columns of the dataset 
print(f"your rows are{df.shape[0]} and your columns are {df.shape[1]}")
```

```python
#now let's see how many null values in total
df.isnull().sum()
```

```python
#which team won the most matches
match_wins=df["match_winner"].value_counts()
match_wins
```

```python
sns.barplot(y=match_wins.index,x=match_wins.values,palette="rainbow")
plt.title("Most match win by Team")
plt.xlabel("Match Wins")
```

```python
#toss decision ttrends
sns.countplot(x=df["toss_decision"],palette="crest")
plt.title("Toss decision Trends")
```

```python
#tosswinner vs matchwinner
count=df[df["toss_winner"]==df["match_winner"]]["match_id"].count()
percentage=count*100/df["match_id"].count()
percentage.round(2)
```

```python
# 2 top top_scorer 
high=df.groupby("top_scorer")["highscore"].sum().sort_values(ascending=False).head(2)
high
high.plot(kind="bar")
```

```python
# top 2 bowling performance
df["best_bowling_figure"]=df["best_bowling_figure"].str.split("--").str[0]
df["best_bowling_figure"]=df["best_bowling_figure"].astype(int)
```

```python
bestbow=df.groupby("best_bowling")["best_bowling_figure"].sum().sort_values(ascending=False).head(10)
bestbow
```

```python
bestbow.plot(kind="bar")
```
